```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 解决中文乱码的问题
plt.rcParams["font.sans-serif"]='SimHei'
# 解决负号无法正常显示的问题
plt.rcParams['axes.unicode_minus']=False
# 让图表直接在Jupyter Notebook中展示出来
%matplotlib inline
# #矢量图设置
%config InlineBackend.figure_format='svg'

# 字体大小
SMALLER = 9
SMALL_SIZE = 12
MEDIUM_SIZE = 14
MEDIUM_SIZE_a = 16
BIGGER_SIZE = 18

font1 = {"family": 'Times New Roman', "style": "normal",
            "variant": "small-caps", "weight": "medium", "size": 17}
font2 = {"family": 'Times New Roman', "style": "normal",
            "variant": "small-caps", "size": 9}
font3 = {"family": 'Times New Roman', "style": "normal",
            "variant": "small-caps", "weight": "bold", "size": 18}
```

```python
# 生成数据
x = ['A', 'B', 'C', 'D', 'E']
y = [100, 200, 300, 400, 500]
filename='ws_mybar'
```

```python
def plot_bar(x,y,filename='ws_mybar'):
    # 创建一行两列的画布
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))

    # xy轴刻度与标签
    for a in ax:
        a.set_xticklabels(x,fontproperties='Palatino Linotype', size=12)

        # 删除上方和右边的边框线，对下方和左边的边框线加粗
        # a.spines['top'].set_color('none')
        # a.spines['right'].set_color('none')
        # bwith = 2  # 边框宽度设置为2
        # a.spines['bottom'].set_linewidth(bwith)
        # a.spines['left'].set_linewidth(bwith)

    ax[0].set_yticklabels(range(0,max(y)+100,100)) # fontproperties='Palatino Linotype'
    ax[0].set_xlabel('Time(ns)',fontsize=16)
    ax[0].set_ylabel('Umax[$m/s$]',fontsize=16)
    ax[1].set_yticklabels(range(100,max(y)+100,100)) # fontproperties='Palatino Linotype'
    ax[1].set_xlabel('Time(us)',fontsize=16)
    ax[1].set_ylabel(r'$\rho$-min [$kg/n^3$]',fontsize=16)

    # 画柱状图
    ax[0].bar(x,y, width=0.8,
         ls='-', lw=1.5,label='Y Distribution') # color='white',ec='black'
    ax[0].set_title('a)', loc='left')
    ax[0].legend(loc='upper left',shadow=False)
    ax[1].barh(x,y,
        ls='-',lw=1.5,label='Y Distribution')# color='white',ec='black'
    ax[1].set_title('b)', loc='left')
    ax[1].legend(loc='lower right', shadow=False)

    # 紧凑式布局
    fig.tight_layout()
    #设置默认字体大小
    plt.rc('font', size=12)          # controls default text sizes
    plt.rc('axes', titlesize=14)    # fontsize of the axes title
    plt.rc('axes', labelsize=18)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
    plt.rc('legend', fontsize=12)       # legend fontsize
    plt.rc('figure', titlesize=16)  # fontsize of the figure title
    # 默认图例的边框
    plt.rcParams['legend.frameon'] = False
    # 默认图例的标记长度
    plt.rcParams['legend.handlelength'] = 2.0
    # 默认标记大小
    plt.rcParams['lines.markersize'] = 5

    # 设置plt的全局参数
    plt.rcParams.update({
     "text.usetex": True,
     "font.family": "san-serif",
     "font.sans-serif": ["Helvetica"]})
    # 保存图片
    plt.savefig(filename+'.tif', format="tif", dpi=400, bbox_inches='tight')
    plt.show()

plot_bar(x,y,filename)
```
